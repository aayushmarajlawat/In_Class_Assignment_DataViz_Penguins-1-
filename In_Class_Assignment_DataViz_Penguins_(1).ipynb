## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="Set2"
)
plt.title("Bill Length vs Bill Depth by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

**Your interpretation (2–3 sentences):**

Each species forms its own cluster of points. Adelie penguins have shorter but deeper bills. Gentoo penguins have longer and shallower bills, and Chinstrap penguins fall in between. This shows that bill size and shape differ clearly among species.


## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
penguins['bill_length_mm'].plot(kind='hist', bins=10, edgecolor='black', title='Histogram with 10 bins')

```

```python
# TODO: Histogram with 30 bins

penguins['bill_length_mm'].plot(kind='hist', bins=30, edgecolor='black', title='Histogram with 30 bins')
```

**Your comparison (1–2 sentences):**

With 10 bins, the histogram is more generalized and smooth, hiding some details of the data distribution. With 30 bins, the histogram shows more variability and reveals finer details, though it appears noisier.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
penguins.boxplot(column='body_mass_g', by='species', grid=False)

```
